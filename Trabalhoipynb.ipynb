Foi escolhido o bando de dados: Rain in Australia: Dados diários de precipitação na Austrália link: https://www.kaggle.com/datasets/jsphyg/weather-dataset-rattle-package

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo diretamente do caminho correto
df = pd.read_csv("weatherAUS.csv")
pd.read_csv("weatherAUS.csv")
```

Aqui foi feito a importação das Bibliotecas e do Banco de Dados


```python
df['Date'] = pd.to_datetime(df['Date'])

df_filtered_date = df[(df['Date'] >= '2010-01-01') & (df['Date'] <= '2015-12-31')]

print(f"Número de registros no intervalo selecionado: {len(df_filtered_date)}")
```

Foi feito a filtragem dos dados para o intervalo de 2010 a 2015


```python
station_name = "Sydney"
df_filtered_station = df_filtered_date[df_filtered_date['Location'] == station_name]

print(f"Registros para a estação {station_name}: {len(df_filtered_station)}")
```

Foi filtrado os dados de uma estação específica, como "Sydney"


```python
df_filtered_station['Date'] = pd.to_datetime(df_filtered_station['Date'])

df_filtered_station['Year'] = df_filtered_station['Date'].dt.year
df_filtered_station['Month'] = df_filtered_station['Date'].dt.month

monthly_avg = df_filtered_station.groupby(['Year', 'Month'])['Rainfall'].mean().reset_index()
print("Médias mensais de precipitação:")
print(monthly_avg.head())
```

Aqui foi feito o cálculo mensal da média de precipitação

```python
plt.figure(figsize=(12, 6))
plt.plot(pd.to_datetime(monthly_avg['Year'].astype(str) + '-' + monthly_avg['Month'].astype(str)),
         monthly_avg['Rainfall'], label="Precipitação Média Mensal")
plt.title("Tendência de Precipitação Mensal (Sydney, 2010-2015)")
plt.xlabel("Tempo")
plt.ylabel("Precipitação (mm)")
plt.legend()
plt.grid()
plt.show()
```

Aqui temos o gráfico de tendências

E como análise do resultado, foi concluído:

*  Tendência de Precipitação: O gráfico mostra flutuações sazonais, com picos geralmente associados a estações chuvosas.
